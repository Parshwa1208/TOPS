```python
#better code 
class ATM:
    def __init__(self):
        print('''Welcome to ATM!
                 0. Press 0 to create pin.
                 1. Press 1 to change pin.
                 2. Press 2 to deposit money.
                 3. Press 3 to check balance.
                 4. Press 4 to withdraw money.
                 5. Press 5 to exit.''')
        
        self.pin = None
        self.balance = 0
        self.menu()

    def menu(self):
        while True:
            try:
                user_input = int(input("\nEnter your choice: "))
                if user_input == 0:
                    self.create_pin()
                elif user_input == 1:
                    self.change_pin()
                elif user_input == 2:
                    self.deposit_money()
                elif user_input == 3:
                    self.check_balance()
                elif user_input == 4:
                    self.withdraw_money()
                elif user_input == 5:
                    print("Thank you for using our ATM. Goodbye!:)))")
                    break
                else:
                    print("Invalid choice. Please try again.")
            except ValueError:
                print("Invalid input! Please enter a number.")

    def create_pin(self):
        self.pin = input("Set a new 4-digit PIN: ")
        print("PIN has been generated successfully.")

    def change_pin(self):
        if self.authenticate():
            new_pin = input("Enter a new 4-digit PIN: ")
            self.pin = new_pin
            print("New PIN has been updated successfully.")

    def deposit_money(self):
        if self.authenticate():
            try:
                amount = float(input("Enter the amount to deposit: "))
                if amount > 0:
                    self.balance += amount
                    print(f"₹{amount} deposited successfully. Current balance: ₹{self.balance}")
                else:
                    print("Deposit amount must be greater than 0.")
            except ValueError:
                print("Invalid amount! Please enter a number.")

    def check_balance(self):
        if self.authenticate():
            print(f"Your current balance is ₹{self.balance}.")

    def withdraw_money(self):
        if self.authenticate():
            try:
                amount = float(input("Enter the amount to withdraw: "))
                if 0 < amount <= self.balance:
                    self.balance -= amount
                    print(f"Withdrawal successful! Remaining balance: ₹{self.balance}")
                else:
                    print("Insufficient funds or invalid amount.")
            except ValueError:
                print("Invalid input! Please enter a number.")

    def authenticate(self):
        entered_pin = input("Enter your PIN: ")
        if entered_pin == self.pin:
            return True
        else:
            print("Incorrect PIN! Yaad Krrrr Bhai.")
            return False



```
